### 4.1 Simulación de valores faltantes

El dataset original no presenta valores faltantes. Sin embargo, se simularon algunos valores nulos en las variables X1, X3 y X7.

Posteriormente se aplicó imputación por media, reemplazando cada valor faltante por el promedio de la respectiva variable.

```python
df_imp = df.copy()

df_imp.loc[10, "X1"] = np.nan
df_imp.loc[50, "X3"] = np.nan
df_imp.loc[100, "X7"] = np.nan
```

```python
df_imp.isnull().sum()
```

### 4.2 Imputación

La imputación se realizó utilizando la media de cada variable.

Tras el proceso, el dataset volvió a presentar cero valores faltantes, quedando apto para las siguientes etapas de modelado.

```python
df_imp["X1"] = df_imp["X1"].fillna(df_imp["X1"].mean())
df_imp["X3"] = df_imp["X3"].fillna(df_imp["X3"].mean())
df_imp["X7"] = df_imp["X7"].fillna(df_imp["X7"].mean())
```

```python
df_imp.isnull().sum()
```

## 7. Normalización / Estandarización (Transformación)

Las variables predictoras presentan escalas diferentes. Por ejemplo, X1 toma valores entre 0.62 y 0.98, mientras que X2 alcanza valores superiores a 800.

De acuerdo con lo explicado en clase, cuando las variables tienen magnitudes muy distintas, algunas pueden influir más que otras durante el entrenamiento de modelos basados en distancia.

Por esta razón se aplicó una estandarización utilizando StandardScaler, transformando las variables de entrada para que tengan media cercana a 0 y desviación estándar cercana a 1.

Las variables objetivo Y1 (Heating Load) y Y2 (Cooling Load) no fueron transformadas debido a que representan los valores que posteriormente se desean predecir.

```python
from sklearn.preprocessing import StandardScaler

X = df_imp.drop(columns=["Y1", "Y2"])

scaler = StandardScaler()

X_scaled = pd.DataFrame(
    scaler.fit_transform(X),
    columns=X.columns
)

X_scaled.head()
```

### Resultado de la estandarización

Se observa que las variables transformadas presentan medias cercanas a 0 y desviaciones estándar cercanas a 1.

Esto confirma que la estandarización fue aplicada correctamente y que las diferencias de escala entre las variables originales fueron eliminadas.

El conjunto de datos queda preparado para algoritmos sensibles a la escala, como K-Means, SVM y redes neuronales.

```python
X_scaled.describe()
```

## 8. Exportar y División del Dataset

El dataset se divide en conjuntos de entrenamiento y prueba utilizando una proporción 80/20.

Esta división permite entrenar modelos posteriormente y evaluar su capacidad de generalización sobre datos no vistos.

```python
df_model = X_scaled.copy()

df_model["Y1"] = df_imp["Y1"]
df_model["Y2"] = df_imp["Y2"]

df_model.head()
```

```python
X = df_model.drop(columns=["Y1", "Y2"])

y = df_model[["Y1", "Y2"]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
df_model.to_csv("../data/processed/data_clean.csv",index=False)

print("Dataset exportado correctamente")
```
